```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
import os
import tqdm
import datetime



json_path = "../taxonomy.json"
with open(json_path, "r") as json_file:
    json_dicts= json.load(json_file)
json_dict = {
    json_dict["common_name"]: json_dict["id"]
    for json_dict in json_dicts["taxonomy"]}

rev_json_dict = {json_dict[k]: k for k in json_dict.keys()}
len(rev_json_dict)

def maximize_value(dictionary):
    # written with the help of Brian :)
    return max((v, k) for (k, v) in dictionary.items())[1]
```

```python
# List all CSV files by name
csv_dir = "/Users/phinchosherpa/Desktop/nyu/BirdVox-300h_csv-annotations/"
csv_names = os.listdir(csv_dir)
        
species_dict=[]
for keys in rev_json_dict:
    species_dict.append(keys)
name=[]

for keys in json_dict:
    name.append(keys) 
    
out_df = pd.DataFrame(columns=["taxon_id", "taxon_name", "csv_name", "minute", "minute_count"])
    
    
for i in tqdm.tqdm(range(22)):
    csv_counts = {}
    for csv_name in csv_names:
        csv_path = os.path.join(csv_dir, csv_name)
        df = pd.read_csv(csv_path)
        booleans = df["Taxonomy Code"].str.startswith(species_dict[i])
        csv_count = sum(booleans)
        csv_counts[csv_name] = csv_count
    best_csv_name = maximize_value(csv_counts)
    best_csv_path = os.path.join(csv_dir, best_csv_name)
    taxon_df = pd.read_csv(best_csv_path)
    taxon_minutes = (taxon_df["Time (s)"]//60).astype("int")
    best_minute, minute_count = Counter(taxon_minutes).most_common()[0]
    row = {
        "taxon_id": species_dict[i],
        "taxon_name": name[i],
        "csv_name": best_csv_name,
        "minute": best_minute,
        "minute_count": minute_count
    }
    out_df = out_df.append(row, ignore_index=True)
        
    
out_df.to_csv("BirdVox-22min.csv")      
```
