```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import math
xval = np.arange(0, 2*np.pi, 0.01)
yval = np.ones_like(xval)

xticks_deg = [80, 200, 310]
xticks_rad = [math.radians(x) for x in xticks_deg]
xticklabels = ["entropy", "spectral\nflux", "phase\ndeviation"]


fig = plt.figure()
colormap = plt.get_cmap('hsv')
norm = mpl.colors.Normalize(0.0, 2*np.pi)

ax = plt.subplot(1, 1, 1, polar=True)
ax.scatter(xval, yval,c=xval, s=300, cmap=colormap,norm=norm,linewidths=0)
ax.set_xticks([])
ax.set_xticklabels([xticklabels[0], xticklabels[1], xticklabels[2]])
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_yticks([])

for xtick_deg in xticks_deg:
    plt.arrow(xtick_deg/180.*np.pi, 0, 0, 1.4,  width = 0.015,
        edgecolor = 'black', facecolor = 'green', lw = 2, zorder = 5)
    
#arr2 = plt.arrow(xticks_deg[1]/180.*np.pi, 0, 0, 1.3,  width = 0.015,
#                edgecolor = 'black', facecolor = 'green', lw = 2, zorder = 5)
#arr2 = plt.arrow(xticks_deg[2]/180.*np.pi, 0, 0, 1.3,  width = 0.015,
#                edgecolor = 'black', facecolor = 'green', lw = 2, zorder = 5)
#ax.AlphaScale = 'log'
#cb.outline.set_visible(False)                                 
#ax.set_axis_off()

ax.spines['polar'].set_visible(False)
```
