```python
%matplotlib inline
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import skimage.io as io
import os
from copy import deepcopy

img_path = '/Users/phinchosherpa/Desktop/nyu/color_spots.jpg'
color = io.imread(img_path)
X = color

R = deepcopy(X);
R[:, :, 1] = 0;
R[:, :, 2] = 0;

G = deepcopy(X);
G[:, :, 0] = 0;
G[:, :, 2] = 0;

B = deepcopy(X);
B[:, :, 0] = 0;
B[:, :, 1] = 0;


C = R + B
M = R + G
Y = B + G


fig, ax = plt.subplots(ncols=2, nrows=2, figsize=(9, 9))
ax[0,0].imshow(color)
ax[0,1].imshow(R)
ax[1,0].imshow(G)
ax[1,1].imshow(B)


fig, ax = plt.subplots(ncols=2, nrows=2, figsize=(9, 9))
ax[0,0].imshow(color)
ax[0,1].imshow(C)
ax[1,0].imshow(M)
ax[1,1].imshow(Y)
```
